```python
from __future__ import print_function
import json

datafile = "/Users/wmoxbury/data/EU petition/131215.json"
tmp = json.load(open(datafile, 'r'))
print(tmp.keys())
print(tmp['links'])

data = tmp['data']
print(data.keys())
```

```python
# the petition structure:

petition = data['attributes']
print(petition['creator_name'])
petition.keys()
```

```python
# how many signatures?

petition['signature_count']
```

```python
# which countries?

countries = petition['signatures_by_country']
for x in sorted(countries, key=lambda x: x['signature_count'], reverse=True):
    print(x['signature_count'], '\t', x['name'])
```

```python
# which constituencies?

constituencies = petition['signatures_by_constituency']
for x in sorted(constituencies, key=lambda x: x['signature_count'], reverse=True):
    print(x['signature_count'], '\t', x['ons_code'], '\t',x['name'])
```

```python
print(sum([x['signature_count'] for x in countries]))
print(sum([x['signature_count'] for x in constituencies]))
```
