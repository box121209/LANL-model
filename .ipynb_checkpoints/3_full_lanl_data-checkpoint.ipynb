# Timings

Cache the timestamp sequences for the three data sets for ocmparison:

```python
proc_times = procs.map(lambda x: x[0])
proc_times.cache()
flow_times = flows.map(lambda x: x[0])
flow_times.cache()
dns_times = dns.map(lambda x: x[0])
dns_times.cache()
```

Take a first look at the three time series of 1-hour event counts over the 58-day period:

```python
def compute_series(binsize = 3600, timerange=1e09):

    proc_count = proc_times.filter(lambda x: x<timerange).map(lambda x: x/binsize).countByValue()
    flow_count = flow_times.filter(lambda x: x<timerange).map(lambda x: x/binsize).countByValue()
    dns_count = dns_times.filter(lambda x: x<timerange).map(lambda x: x/binsize).countByValue()
    
    return [proc_count, flow_count, dns_count]

def show_series(pc, fc, dc):
    
    plt.figure(figsize=(20,8))
    ts1, = plt.semilogy(pc.keys(), pc.values(), label='processes')
    ts2, = plt.semilogy(fc.keys(), fc.values(), label='flows')
    ts3, = plt.semilogy(dc.keys(), dc.values(), label='DNS requests')
    plt.legend(handles=[ts1, ts2, ts3], loc='lower left')
    plt.show()

[pc, fc, dc] = compute_series()
show_series(pc, fc, dc)
```

Two things stand out immediately: first, the clear daily and weekly cycle in the process time series.

But second, something bad happens after 15 days: the spike in process activity initiates both a catastrophic drop in collection for the netflow data, and a curious re-setting of DNS volumes which eventually settle a factor of 10 higher than the initial period.

We'll try to understand what happens here later - first we want ot check that the timestamps are properly synchronised across the three data sets, by examining the correlation over the good fortnight.

```python
# the initial fortnight in 5 minute bins:

tr = 370*binsize

[pc_init, fc_init, dc_init] = compute_series(binsize = 300, timerange = tr)
show_series(pc_init, fc_init, dc_init)

print("Time range:", tr/3600.0/24.0, "days")
```

This initial 15 day period is where the 'clean' data appears to be. It's very interesting to view the three mutual correlation plots, which exhibit clean daily and weekly cycles:

```python
# examine correlation coefficients for lagging up to one week:

from scipy import stats

x = pc_init.values()
y = fc_init.values()
z = dc_init.values()

f = stats.spearmanr  # monotonic correlation
#f = stats.pearsonr  # linear correlation

n = min(len(x), len(y), len(z))
corr_xy = [f(x[:(n - offset)],y[offset:n])[0] for offset in range(7*12*24)]
corr_yz = [f(y[:(n - offset)],z[offset:n])[0] for offset in range(7*12*24)]
corr_zx = [f(z[:(n - offset)],x[offset:n])[0] for offset in range(7*12*24)]

plt.figure(figsize=(20,8))
plt.plot(corr_xy)
plt.plot(corr_yz)
plt.plot(corr_zx)
plt.show()
```

I can't imagine a cleaner demonstration of the 7-day weekly structure in the LANL data!

```python
# the critical 11-hour period:

start_hour = 339
end_hour = 350
x = pc.keys()[start_hour:end_hour]
y = pc.values()[start_hour:end_hour]
plt.plot(x, y)
```

```python
# the event sets for this period:

proc_chg = procs.filter(lambda x: x[0]>3600*start_hour and x[0]<3600*end_hour)
flows_chg = flows.filter(lambda x: x[0]>3600*start_hour and x[0]<3600*end_hour)
dns_chg = dns.filter(lambda x: x[0]>3600*start_hour and x[0]<3600*end_hour)
proc_chg.cache()
flows_chg.cache()
dns_chg.cache()
```

```python
# the time series for this period at 1-minute resolution:

binsize = 60
proc_ts = proc_chg.map(lambda x: x[0]).map(lambda x: x/binsize).countByValue()
flow_ts = flows_chg.map(lambda x: x[0]).map(lambda x: x/binsize).countByValue()
dns_ts = dns_chg.map(lambda x: x[0]).map(lambda x: x/binsize).countByValue()

# these need to be sorted:
from collections import OrderedDict
proc_ts = OrderedDict(sorted(proc_ts.items()))
flow_ts = OrderedDict(sorted(flow_ts.items()))
dns_ts = OrderedDict(sorted(dns_ts.items()))

# now plot:
plt.figure(figsize=(20,8))
ts1, = plt.semilogy(proc_ts.keys(), proc_ts.values(), label='processes')
ts2, = plt.semilogy(flow_ts.keys(), flow_ts.values(), label='flows')
ts3, = plt.semilogy(dns_ts.keys(), dns_ts.values(), label='DNS requests')
plt.legend(handles=[ts1, ts2, ts3], loc='lower left')
plt.show()
```
